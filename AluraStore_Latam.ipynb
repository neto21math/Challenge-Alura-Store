### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# INGRESO POR TIENDA
ingreso_tienda1 = tienda['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Mostrar los ingresos totales
print(f"Ingreso total Tienda 1: {ingreso_tienda1}")
print(f"Ingreso total Tienda 2: {ingreso_tienda2}")
print(f"Ingreso total Tienda 3: {ingreso_tienda3}")
print(f"Ingreso total Tienda 4: {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
# Función para calcular y mostrar las ventas por categoría para una tienda
def ventas_por_categoria(df, nombre_tienda):
    ventas_categoria = df.groupby('Categoría del Producto').size().reset_index(name='Cantidad de Ventas')
    ventas_categoria_ordenado = ventas_categoria.sort_values(by='Cantidad de Ventas', ascending=False)
    print(f"Ventas por categoría en {nombre_tienda}:")
    display(ventas_categoria_ordenado)
    print("\n")

# Calcular y mostrar ventas por categoría para cada tienda
ventas_por_categoria(tienda, "Tienda 1")
ventas_por_categoria(tienda2, "Tienda 2")
ventas_por_categoria(tienda3, "Tienda 3")
ventas_por_categoria(tienda4, "Tienda 4")
```

# 3. Calificación promedio de la tienda


```python
# Calcular la calificación promedio para cada tienda
calificacion_tienda1 = tienda['Calificación'].mean()
calificacion_tienda2 = tienda2['Calificación'].mean()
calificacion_tienda3 = tienda3['Calificación'].mean()
calificacion_tienda4 = tienda4['Calificación'].mean()

# Mostrar las calificaciones promedio
print(f"Calificación promedio Tienda 1: {calificacion_tienda1}")
print(f"Calificación promedio Tienda 2: {calificacion_tienda2}")
print(f"Calificación promedio Tienda 3: {calificacion_tienda3}")
print(f"Calificación promedio Tienda 4: {calificacion_tienda4}")
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función para identificar y visualizar los productos más y menos vendidos
def productos_vendidos(df, nombre_tienda):
    ventas_producto = df['Producto'].value_counts()
    productos_mas_vendidos = ventas_producto.head(10)
    productos_menos_vendidos = ventas_producto.tail(10)

    print(f"Productos más vendidos en {nombre_tienda}:")
    display(productos_mas_vendidos)
    print("\n")

    print(f"Productos menos vendidos en {nombre_tienda}:")
    display(productos_menos_vendidos)
    print("\n")



# Identificar y visualizar productos más y menos vendidos para cada tienda
productos_vendidos(tienda, "Tienda 1")
productos_vendidos(tienda2, "Tienda 2")
productos_vendidos(tienda3, "Tienda 3")
productos_vendidos(tienda4, "Tienda 4")
```

# 5. Envío promedio por tienda

```python
# Calcular el costo de envío promedio para cada tienda
envio_promedio_tienda1 = tienda['Costo de envío'].mean()
envio_promedio_tienda2 = tienda2['Costo de envío'].mean()
envio_promedio_tienda3 = tienda3['Costo de envío'].mean()
envio_promedio_tienda4 = tienda4['Costo de envío'].mean()

# Mostrar los costos de envío promedio
print(f"Costo de envío promedio Tienda 1: {envio_promedio_tienda1}")
print(f"Costo de envío promedio Tienda 2: {envio_promedio_tienda2}")
print(f"Costo de envío promedio Tienda 3: {envio_promedio_tienda3}")
print(f"Costo de envío promedio Tienda 4: {envio_promedio_tienda4}")
```

## Recomendación Final: Venta de la Tienda 4

El Sr. Juan debe vender o liquidar los activos de la **Tienda 4** para obtener capital e iniciar su nuevo negocio.

### Justificación de la Venta (Tienda 4)

La Tienda 4 es el activo de menor valor y mayor riesgo:

1.  **Riesgo de Mercado (Tendencia Crítica):** Registró el **ingreso total más bajo** (\$1,038,375,700.0) y, de forma crítica, sus **ventas mensuales están en declive**. Es un mercado en contracción, lo que justifica su liquidación.
2.  **Bajo Valor Estratégico:** A pesar de su bajo coste de envío, la combinación de bajo volumen y calificación de 3.996 la convierte en la opción menos atractiva para el futuro.

### Justificación de las Tiendas a Conservar (1, 2 y 3)

El Sr. Juan debe conservar las Tiendas 1, 2 y 3, ya que representan una cartera de alto rendimiento, estabilidad y calidad:

* **Tienda 1 (Volumen):** Se conserva por su **liderazgo irrefutable en Ingresos Totales** (\$1,150,880,400.0). Es el canal de mayor alcance y debe ser el foco principal de inversión.
* **Tienda 3 (Eficiencia y Calidad):** Se conserva por su **máxima calificación promedio** (4.048) y su **eficiencia logística** (el menor coste de envío entre los canales de alto volumen).
* **Tienda 2 (Estabilidad Intermedia):** Se conserva porque, a pesar de ser la segunda en ingresos (\$1,116,343,500.0), mantiene una **sólida posición financiera** en el medio y contribuye a la diversificación de canales sin el riesgo de inestabilidad de la Tienda 4.

**Conclusión:** La venta de la Tienda 4 elimina el riesgo inminente de pérdidas, permitiendo al Sr. Juan capitalizar su negocio menos viable para financiar el próximo proyecto.
